# Data

```python
from llama_index.readers.web import BeautifulSoupWebReader

urls = [
    "https://raw.githubusercontent.com/mieweb/docs/master/content/functions/document-management.md",
    "https://raw.githubusercontent.com/mieweb/docs/master/content/functions/e-chart.md",
    "https://raw.githubusercontent.com/mieweb/docs/master/content/functions/e-sign-electronic-signature.md",
    "https://raw.githubusercontent.com/mieweb/docs/master/content/functions/encounters.md",
    "https://raw.githubusercontent.com/mieweb/docs/master/content/functions/injury-care.md",
    "https://raw.githubusercontent.com/mieweb/docs/master/content/functions/patient-registration.md"
]

web_reader = BeautifulSoupWebReader()

documents = []
for url in urls:
    documents.extend(web_reader.load_data([url]))

for doc in documents:
    print(doc)
```

```python
for doc in documents:
    print(doc)
```

# LLM

```python
from llama_index.llms.groq import Groq

from google.colab import userdata
llm = Groq(model="llama3-8b-8192", api_key=userdata.get('GROQ_API_KEY'))
```

# Embedding Model

```python
from llama_index.core import Settings
from llama_index.embeddings.huggingface import HuggingFaceEmbedding

Settings.llm = llm
Settings.embed_model = HuggingFaceEmbedding(
    model_name="BAAI/bge-small-en-v1.5"
)
```

# Index Setup

```python
from llama_index.core import VectorStoreIndex

vector_index = VectorStoreIndex.from_documents(documents)
```

```python
from llama_index.core import SummaryIndex

summary_index = SummaryIndex.from_documents(documents)
```

## Router Query Engine

```python
from llama_index.core.tools import QueryEngineTool, ToolMetadata

vector_tool = QueryEngineTool(
    vector_index.as_query_engine(),
    metadata=ToolMetadata(
        name="vector_search",
        description="Useful for searching for specific facts.",
    ),
)

summary_tool = QueryEngineTool(
    summary_index.as_query_engine(response_mode="tree_summarize", verbose=True), # The verbose argument should be passed here
    metadata=ToolMetadata(
        name="summary",
        description="Useful for summarizing an entire document.",
    ),
)

tools = [vector_tool, summary_tool]
```

```python
Settings.debug=True
```

```python
from llama_index.core.query_engine import RouterQueryEngine

query_engine = RouterQueryEngine.from_defaults(
    [vector_tool, summary_tool], select_multi=False, verbose=True,

)

response = query_engine.query("What was mentioned about E-chart?")

display_response(response)
```

```python
response = query_engine.query(
    "What was mentioned about patient care? What other tools are mentioned and what was said about them?"
)

display_response(response)
```

#### Multi Selector

```python
from llama_index.core.query_engine import RouterQueryEngine

query_engine = RouterQueryEngine.from_defaults(
    [vector_tool, summary_tool],
    select_multi=True,
    verbose=True,
)

response = query_engine.query(
    "What was mentioned about Registration system? Summarize what is mentioned about other mangement in the document."
)

display_response(response)
```
